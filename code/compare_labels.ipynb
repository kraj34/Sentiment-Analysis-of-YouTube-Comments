```python
import pandas as pd
```

```python
llm_df = pd.read_csv('llm_labeled_data.csv')
bert_df = pd.read_csv('youtube_hello_comments_clustered.csv')
```

```python
llm_df = llm_df.dropna()
llm_df
```

```python
bert_df = bert_df[['comment', 'sentiment']]
bert_df = bert_df.rename(columns={'comment': 'Comment', 'sentiment': 'Label'})
bert_df['Label'] = bert_df['Label'].str.lower()
bert_df
```

```python
merged = pd.merge(llm_df, bert_df, on='Comment', how='outer')
merged = merged.drop_duplicates(subset='Comment', keep='first')
merged = merged.rename(columns={'Label_x': 'LLM_Label', 'Label_y': 'BERT_Label'})
merged
```

```python
merged[['LLM_Label', 'BERT_Label']].describe()
```

```python
print(merged['LLM_Label'].value_counts())
merged['BERT_Label'].value_counts()
```

```python
ac = (merged.LLM_Label == merged.BERT_Label).mean()
print(f'Agreement between LLM and BERT: {ac*100:.2f}%')
```
